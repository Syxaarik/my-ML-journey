## Устанавливаем либы

```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

## Представнение вектора


```python
# начальная точка
X, Y = [0], [0]

# напровление вектора, его конечная точка
U, V = [2], [2]

# инициализирует график: fig - холст на котором распологаюстя графики; ax - само координатное поле
fig, ax = plt.subplots(figsize=(6,6))

# Отрисовка координатных линий x, -x, y, -y через центр
ax.axhline(0, color='black', linewidth=1.2)  # Горизонтальная ось
ax.axvline(0, color='black', linewidth=1.2)  # Вертикальная ось

# Отрисовка вектора из центра
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='blue', zorder=3)

# Смещаем вектор в нижний левый угол, настраивая видимые границы
# Центр (0,0) будет смещен вниз и влево
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])

# Настройка сетки и внешнего вида
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_aspect('equal')  # Чтобы масштаб X и Y был одинаковым

plt.show()
```

```python
# задаем матрицу линейного преобразования 
A = np.array([[0, 2],
              [2, 2]])

# задаем базисный вектор 
i = np.array([1, 0])
j = np.array([0, 1])

# применяем преобразования
i_transform = A @ i
j_transform = A @ j

fig, ax = plt.subplots(figsize=(6, 6))

# Функция для удобного рисования стрелок векторов
def draw_vector(v, color, label):
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)

# Рисуем исходные векторы (пунктиром или бледным цветом)
draw_vector(i, 'lightblue', 'Исходный i (1,0)')
draw_vector(j, 'lightcoral', 'Исходный j (0,1)')

# Рисуем трансформированные векторы
draw_vector(i_transform, 'blue', 'Новый i')
draw_vector(j_transform, 'red', 'Новый j')

# Оформление сетки и осей
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
ax.grid(True, linestyle='--')
ax.set_aspect('equal') # Важно, чтобы масштаб осей X и Y был одинаковым
plt.legend()
plt.title("Линейное преобразование векторов")
plt.show()
```
